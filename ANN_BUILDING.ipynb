```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled = StandardScaler()
X_train = scaled.fit_transform(X_train)
X_test = scaled.transform(X_test)
```

```python
import keras
```

```python
#The sequential model reqd. to initialize our NN and the dense 
#module required to build the layers of our NN
```

```python
#seq. module
from keras.models import Sequential
#Dense
from keras.layers import Dense
#Drop out
#you can apply dropout to 1 or more layers
from keras.layers import Dropout
```

```python
#Initializing NN
#2 ways :-
#Either by defining the seq. of layers or by defining a graph

classifier = Sequential()
```

```python
#Adding input layer and first hidden layer
#step 1 : Initialize wts. to small nos..The dense function in the 
#dense module takes care of this
#Step 2 : Each feture in one input node. No. of nodes in i/p layer=
#no. of params.
#Step 3 : 

#classifier.add(layer) - we use add method to add the different 
#layers in our neural network
#In the dense paranthesis we are going to input the params. in the
#steps involved ,i.e., how the wts. are dated,the activatn. functn.
#we use and what is the no. of nodes in that layer and no. of input
#nodes

#The no. of nodes in the i/p layer = 11 (no. of params.)
#The no. of nodes in o/p layer = 1 (0 or 1)
#kernel_initializer arg. is used to initialize wt.
classifier.add(Dense(units=6,kernel_initializer = 'uniform',activation='relu',input_shape=(11,)))
#Applying dropout
#rate arg. - fraction of the neurons you want to drop out
#rate=0.1 means 10% of neurons will be disabled for each iteration.
classifier.add(Dropout(rate=0.1))
```

```python
#Adding the second hidden layer

classifier.add(Dense(units=6,kernel_initializer = 'uniform',activation='relu'))
classifier.add(Dropout(rate=0.1))
```

```python
#Adding the output layer

#As we only have 2 categories for o/p (0 or 1) we chose 'sigmoid'
#if more than 2 categories the use 'softmax'
classifier.add(Dense(units=1,kernel_initializer = 'uniform',activation='sigmoid'))
```

```python
#Compiling our NN
#Basically applying stochastic G.D. on the whole NN

#optimizer arg. - simply the algo u want to use to find the optimal
#set of wts..There are several types of stochastic G.D. the most 
#efficient is 'Adam'

#loss arg. - corresponds to the loss function within the stochastic
#G.D. algo i.e. within the Adam algo.

#if your D.V. has binary outcome then loss function is
#'binary_crossentropy' else 'categorical_crossentropy'
classifier.compile(optimizer='Adam',loss='binary_crossentropy',metrics= ['accuracy'])
```

```python
#Fitting our NN to training set

classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
#Predicting on the test set

y_pred = classifier.predict(X_test)
```

```python
#Validation
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
# k FOLD CROSS VALIDATION
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
#The keras classifier expects one of its args here to be a function
```

```python
def build_classifier():
    #Here we have to take all the lines of code to build the 
    #artificial NN
    classifier = Sequential()
    
    classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_shape=(11,)))
   
    classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
    
    classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
                   
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    
    return classifier

```

```python
#The classifier var. above is a local var. defined only for above
#function
#Now we have to make a global classifier var.

#
classifier = KerasClassifier(build_fn=build_classifier,batch_size=10,epochs=100)
#accracies contains all the accuracies of our k fold 
#cv arg. - k
accuracies = cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=10)
```

```python
#We have 2 types of params.
#1. That are learnt from the model during the training(like wts.)
#2. That are fixed (hyper parameters)
#It consists of finding the best values of the hyper parameters and 
#it is done with the help of grid search
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    #Here we have to take all the lines of code to build the 
    #artificial NN
    classifier = Sequential()
    
    classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_shape=(11,)))
   
    classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
    
    classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
                   
    classifier.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    
    return classifier
```

```python
classifier = KerasClassifier(build_fn=build_classifier)
```

```python
#rmsprop optimizer is preferred for RNN
param = {'batch_size':[25,32],'epochs':[100,500],
         'optimizer':['adam','rmsprop']}
```

```python
grid = GridSearchCV(estimator=classifier,param_grid=param,
                    scoring='accuracy',cv = 10)
```

```python
grid = grid.fit(X_train,y_train)
best_score = grid.best_score_
```
